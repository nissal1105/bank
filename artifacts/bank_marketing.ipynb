# Step 6. Model Training

```python
x = df.drop('y', axis=1)
y = df['y']

x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=9)
```

```python
x_train.shape,y_train.shape, x_test.shape, y_test.shape
```

```python
#model object

logistic_model = LogisticRegression(max_iter=300)
```

```python
#fitting

logistic_model.fit(x_train,y_train)
```

```python
##prediction

y_pred_logistic = logistic_model.predict(x_test)
y_pred_logistic.shape
```

```python
print("Accuracy of logistic_model",accuracy_score(y_test,y_pred_logistic))
# 0.9073399692481994 @100iter
# 0.9104960750991341 @200iter
# 0.9108197782633325 @300iter

# 0.9123573682932751
```
